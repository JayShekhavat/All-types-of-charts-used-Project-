##**Import some libraries**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

##**1. Line Plo**t

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)', color='blue', linewidth=2)
plt.title('Line Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

##We used line plot to show 'sin' values on x values as an input.

##**2. Scatter Plot**

```python
np.random.seed(0)
data = pd.DataFrame({'X': np.random.rand(100), 'Y': np.random.rand(100)})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='X', y='Y', color='red')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

##Above graph show scatter plot of **X** and **Y** data.

##**3. Bar Plot**

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='purple')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

##This graph show values of each input but **'B'** has high value.

##**4. Histogram**

```python
data = np.random.normal(loc=0, scale=1, size=500)

plt.figure(figsize=(8, 6))
sns.histplot(data, bins=20, kde=True, color='orange')
plt.title('Histogram with KDE')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

###This chart show normally distribution of data

##**5. Box Plot**

```python
data = np.random.normal(0, 1, 100)

plt.figure(figsize=(6, 6))
sns.boxplot(data=data, color='cyan')
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```

###Above graph is boxplot which show **'MIN'**, **'MEAN'**, **'IQR**' and **'MAX'** of each column.

##**6. Violin Plot**

```python
data = pd.DataFrame({
    'Category': ['A'] * 50 + ['B'] * 50,
    'Values': np.concatenate([np.random.normal(0, 1, 50), np.random.normal(2, 1, 50)])
})

plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='Category', y='Values', palette='Set2')
plt.title('Violin Plot')
plt.show()
```

##**Correlation of each columns**

```python
matrix1 = np.random.normal(1, 2, 50)
matrix2 = np.random.rand(5, 5)
print(matrix1)
print(matrix2)
```

##**7. Heatmap**

```python
plt.figure(figsize=(6, 5))
sns.heatmap(matrix2,  annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

##**8. Pair Plot**

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species')
plt.show()
```

```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species', palette='husl')
plt.show()
```

##**9. Pie Chart**

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

##**10. Swarm Plot**

```python
data = sns.load_dataset('tips')

plt.figure(figsize=(8, 6))
sns.swarmplot(data=data, x='day', y='total_bill', hue='sex', palette='deep')
plt.title('Swarm Plot')
plt.show()
```

##**11. Strip Plot**

```python
plt.figure(figsize=(8, 6))
sns.stripplot(data=data, x='day', y='total_bill', hue='sex', palette='Set1', dodge=True)
plt.title('Strip Plot')
plt.show()
```

##**12. Density Plot**

```python
data = np.random.normal(0, 1, 500)

plt.figure(figsize=(8, 6))
sns.kdeplot(data, shade=True, color='blue')
plt.title('Density Plot')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()
```

##**13. Joint Plot**

```python
data = pd.DataFrame({'X': np.random.rand(100), 'Y': np.random.rand(100)})
sns.jointplot(data=data, x='X', y='Y', kind='scatter',  palette='coolwarm')
plt.show()
```

##**14. Count Plot**

```python
data = pd.DataFrame({'X': np.random.rand(100), 'Y': np.random.rand(100)})
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='X', palette='Set3')
plt.title('Count Plot')
plt.show()
```
